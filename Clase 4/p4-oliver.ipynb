1. Implementar un pipeline de clasificación con la muestra que se vió en clase. El código para traer el dataset es el siguiente:

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
print(mnist.keys())
X, y = mnist["data"], mnist["target"]
```

```python
import numpy as np
# Casteamos a int
y = np.array([int(n) for n in y])

print(y)
```

```python
# Escalamos todos los datos a 0 1 (Normalización)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = scaler.fit_transform(X)
```

```python
# Vamos a modelar la detección para el numero 5

from sklearn.model_selection import train_test_split

#Primero vamos a crear los dataset de evaluación y entrenamiento
X_train_set, X_test_set = train_test_split(X, test_size=0.2, random_state=42)

y_train_set, y_test_set = train_test_split(y, test_size=0.2, random_state=42)

print(X_train_set)

print(X_test_set)
```

```python
# Filtramos las etiquetas solo para los valores 5
y_train_set_5 = (y_train_set == 5)

y_test_set_5 = (y_test_set == 5)

print(list(filter(lambda X: X is True, y_test_set_5.tolist())))
```

```python
# Importamos el clasificado binario
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)

sgd_clf.fit(X_train_set, y_train_set_5)
```

```python
# Probamos el modelo
sgd_clf.predict([X[11]])
```

```python
# Midiendo precisión usando validación cruzada
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, X_train_set, y_train_set_5, cv=3, scoring = "accuracy")
```

```python
# Medimos la precisión usando el "dummy classifier"

from sklearn.dummy import DummyClassifier

never_5_clf = DummyClassifier(strategy="most_frequent")

# volvemos a medir la precisión usando validación cruzada

cross_val_score(never_5_clf, X_train_set, y_train_set_5, cv=3, scoring = "accuracy")
```

```python
# Matriz de confusión
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

y_train_pred = cross_val_predict(sgd_clf, X_train_set, y_train_set_5, cv=3)

confusion_matrix(y_train_set_5, y_train_pred)

# Los arroja en el siguiente orden: Verdadero negativo, verdadero positivo, Falso negativo, Falso positivo
```

```python
# Implementación de precisión (V.P / V.P + F.P)
from sklearn.metrics import precision_score, recall_score

precision = precision_score(y_train_set_5, y_train_pred)

recall = recall_score(y_train_set_5, y_train_pred)

print("Precision: {}".format(precision))

print("Recall: {}".format(recall))
```

```python
# Medición del F1 score

from sklearn.metrics import f1_score

f1 = f1_score(y_train_set_5, y_train_pred)

print(f1)
```
